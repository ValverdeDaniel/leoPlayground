```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.uniform(0, 1, size=(25, 9)))

df[1] = df[1].round(2)
df[2] = 'a'
df[3] = df[3] * 100
pd.options.display.float_format = '{:,}'.format
df[4] = df[4] * 100
display(df)
```
