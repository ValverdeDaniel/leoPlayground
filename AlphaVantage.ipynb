```python
from fredapi import Fred
import pandas as pd

fred_key = '098379ebf47578c47d80f7becf90cb34'
# Replace YOUR_API_KEY with your actual API key
fred = Fred(api_key=fred_key)

#Retrieve the data for the EBT
ebt = pd.DataFrame(fred.get_series('EBT'))

#Retrieve the data for the TCI
tci = pd.DataFrame(fred.get_series('TCI'))

#Retrieve the data for the Shares Outstanding
so = pd.DataFrame(fred.get_series('SO'))

#Filter the data for the ticker AAPL
ebt = ebt[ebt.index == 'AAPL']
tci = tci[tci.index == 'AAPL']
so = so[so.index == 'AAPL']

#Calculate the EPS
eps = (ebt - tci) / (so*1000)

#Print the data
print(eps)
```
